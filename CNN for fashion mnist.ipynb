```
import tensorflow as tf
import numpy as np
import pandas as pd
import functools
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Flatten,MaxPool2D,Dense
```

```
fashion_mnist=tf.keras.datasets.fashion_mnist
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```
# display the image
i=0
plt.imshow(x_train[i])
plt.show()
```

```
labels = [
    'T-shirt/top',
    'Trouser',
    'Pullover',
    'Dress',
    'Coat',
    'Sandal',
    'Shirt',
    'Sneaker',
    'Bag',
    'Ankle boot'
]
```

```
scale_x_train=x_train/255
scale_x_test=x_test/255
scale_x_train=scale_x_train[...,np.newaxis]
scale_x_test=scale_x_test[...,np.newaxis]
model=Sequential([
                  Conv2D(filters=8,kernel_size=(3,3),strides=(1,1),padding="SAME",activation="relu",input_shape=scale_x_train[0].shape),
                  MaxPool2D(pool_size=(3,3)),
                  Flatten(),
                  Dense(units=16,activation="relu"),
                  Dense(units=16,activation="relu"),
                  Dense(units=16,activation="softmax"),
])
```

```
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```
details=model.fit(scale_x_train,y_train,batch_size=100,epochs=5,verbose=2)
```

```
frame=pd.DataFrame(details.history)
print(frame)
```

```
frame1=frame.plot(y="loss",title="Epocs vs loss",)
frame1.set(xlabel="Epocs",ylabel="loss")
```

```
frame2=frame.plot(y="accuracy",title="Epocs vs Accuracy")
frame2.set(xlabel="Epocs",ylabel="accuracy")
```

```
loss,acc=model.evaluate(scale_x_test,y_test,verbose=2)
print(f"Loss={loss}")
print(f"Accuracy={acc}")
```
